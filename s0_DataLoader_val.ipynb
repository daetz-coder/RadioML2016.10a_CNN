```python
import pandas as pd

# 指定CSV文件路径
csv_file_path = 'snr_data/output_data_snr_6.csv'  # 替换为你的CSV文件路径

# 加载CSV文件
data_frame = pd.read_csv(csv_file_path)

# 提取前256列数据
vectors = data_frame.iloc[:, :256].values

# 提取Mod_Type列（倒数第二列）
mod_types = data_frame['Mod_Type'].values

# 创建一个字典，用于存储每种类型的数据
type_dict = {}

# 遍历每个Mod_Type，将对应的向量存储到字典中
for mod_type in set(mod_types):
    type_dict[mod_type] = vectors[mod_types == mod_type]

# 打印每种类型的数据数量
for mod_type, data in type_dict.items():
    print(f"Mod_Type: {mod_type}, Number of samples: {data.shape[0]}")
```
